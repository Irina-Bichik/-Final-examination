```python
n_A = df[df['ab_group']=='A']['purchase'].sum()
print (n_A)  # колличество платящих пользователей в группе А
```

```python
n_B = df[df['ab_group']=='B']['purchase'].sum()
print (n_B)   # колличество платящих пользователей в группе В
```

```python
r_A = df[(df['ab_group'] == 'A')&(df['purchase'] == 1)]['revenue'].sum()
print (r_A)   # доход по группе А
```

```python
r_B = df[(df['ab_group'] == 'B')&(df['purchase'] == 1)]['revenue'].sum()
print (r_B)   # доход по группе B
```

```python
ARPPU_А = r_A/n_A
ARPPU_B = r_B/n_B
print (f'ARPPU группы А: {round(ARPPU_А,2)}')
print (f'ARPPU группы B: {round(ARPPU_B,2)}')
```
